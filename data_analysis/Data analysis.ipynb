```python
import math
from copy import deepcopy
```

```python
def calculate_age(age_str):
    if type(age_str) == float:
        return age_str
    elif '-' in age_str:
        ages = age_str.split('-')
        return (int(ages[1]) + int(ages[0]))/2
    elif '+' in age_str:
        return int(age_str.split('+')[0])
```

```python
df_aged = deepcopy(df)
df_aged['Age'] = df_aged['Age'].map(calculate_age)
df_aged['Age']
```

```python
def calculate_income(income_str):
    if type(income_str) == float:
        return income_str
    elif income_str == 'Prefer not to answer':
        return float('nan')
    elif 'to' in income_str:
        incomes_str = income_str.replace(',', '')
        incomes = incomes_str.split(' to ')
        return (int(incomes[1][1:]) + int(incomes[0][1:]))/2
    elif 'and up' in income_str:
        incomes = income_str.replace(',', '')
        return int(incomes.split(' and up')[0][1:])
```

```python
df_income = deepcopy(df)
df_income['income'] = df_income['How much total combined money did all members of your HOUSEHOLD earn last year?'].map(calculate_income)
df_income['income']
```

```python
sauce = 'What type of cranberry saucedo you typically have?'

sauces_data = df_income.groupby([sauce])['income'].mean()
sauces_data.plot.bar()
```

```python
t = pd.pivot_table(df_aged, values='RespondentID', index='Age', columns=sauce, aggfunc=pd.Series.count)
t
```

```python
sauced_ages = {}
for r in t.iteritems():
    for sauce_type in r[1:]:
        sauced_ages[sauce_type.name] = {}
        for age, count in sauce_type.iteritems():
            sauced_ages[sauce_type.name][age] = count
```

```python
for key, value in sauced_ages.items():
    _, ax = plt.subplots()
    ax.set(title='Sauce: {}'.format(key),
           xlabel='age', ylabel='count')
    plt.bar(value.keys(), value.values(), 5)
    plt.show()
```
